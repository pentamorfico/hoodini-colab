# Hoodini Colab — Development Notebook

This notebook is for local development and testing. It aligns installs to the current kernel Python, uses the local repo in editable mode, and forces a module reload so changes to `src/hoodini_colab/widget.js` are reflected immediately.

- Prints kernel Python path and resolved module path
- Uninstalls any `hoodini-colab` from the kernel environment
- Installs the local repo in editable mode
- Reloads modules and launches the widget

```python
# Align installs to the exact kernel Python and use local repo
import subprocess
import sys

print("Kernel Python:", sys.executable)

# Uninstall any hoodini-colab in this kernel's environment
try:
    subprocess.run([sys.executable, '-m', 'pip', 'uninstall', '-y', 'hoodini-colab'], check=True)
except Exception as e:
    print('Uninstall warning:', e)

# Install local repo (editable) so widget.js changes are used
subprocess.run([sys.executable, '-m', 'pip', 'install', '-e', '/home/pentamorfico/software/hoodini-colab'], check=True)

# Hard reload modules to avoid cache
for m in list(sys.modules):
    if m.startswith('hoodini_colab'):
        del sys.modules[m]

from pathlib import Path

import hoodini_colab as hc
from hoodini_colab import create_launcher

print('hoodini_colab from:', hc.__file__)
print('Widget.js path:', Path(hc.__file__).parent / 'widget.js')

# Create and display the interactive launcher
launcher = create_launcher()
display(launcher)


launcher.keep_alive()
```
